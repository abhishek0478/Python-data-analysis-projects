```python
path = '/content/drive/MyDrive/Country-Code.xlsx'
```

```python
df_country = pd.read_excel(path)
```

```python
df_country.head()
```

```python
df_country.shape
```

```python
df1 = pd.merge(df,df_country,on='Country Code', how ='left')
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1.columns
```

```python
country_names = df1.Country.value_counts().index
```

```python
country_values = df1.Country.value_counts().values
```

```python
## Pie Chart- Top 3 countries that uses zomato
plt.pie(country_values[:3], labels=country_names[:3],autopct='%1.2f%%')
```

```python
df1.duplicated().sum()  
```

```python
ratings = df1.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
ratings = ratings.rename(columns={0:'Rating count'})
ratings
```

```python
# plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',hue = 'Rating color',data = ratings, palette = ['blue','red','orange','yellow','green','green'])
```

```python
#Countries with 0 rating
df1.groupby(['Aggregate rating','Country']).size().reset_index().head(5)
```

```python
# find name of currency used by different countries
df1[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

```python
# which countries do have online deliveries options.
df1[df1['Has Online delivery']=='Yes'].Country.value_counts()
```

```python
df1[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```
