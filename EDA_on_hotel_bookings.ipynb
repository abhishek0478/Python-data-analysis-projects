#**Where do the guest come from** ?

```python
country_of_guest = cleaned_data[cleaned_data['is_canceled']==0 ] ['country'].value_counts().reset_index()
country_of_guest.columns=['country','Number of guest']
country_of_guest
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
import plotly.express as px
```

```python
valid_guest = cleaned_data.loc[cleaned_data["is_canceled"]==0]
prices = valid_guest[["hotel","reserved_room_type","adr_updated"]].sort_values("reserved_room_type")
prices
```

```python
cleaned_data["adr_updated"]=cleaned_data["adr"]/(cleaned_data["adults"]+cleaned_data["children"])
valid_guest = cleaned_data.loc[cleaned_data["is_canceled"]==0]
prices = valid_guest[["hotel","reserved_room_type","adr_updated"]].sort_values("reserved_room_type")

plt.figure(figsize=(12,8))
sns.boxplot(x="reserved_room_type",
            y="adr_updated",
            hue = "hotel",
            data = prices
            )
plt.title("price of room types per night and person", fontsize = 16)
plt.xlabel("Room type", fontsize=16)
plt.ylabel("price[EUR]",fontsize=16)

plt.ylim(0,160)
plt.show()
```
